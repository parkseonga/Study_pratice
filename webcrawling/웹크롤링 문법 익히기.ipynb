##### get: 서버로부터 정보를 조회하기 위해 설계된 메소드
* 요청라인과 요청 헤드를 보내야햄
* 요청라인: 요청방식, 경로, HTTP
* 검색 또는 브라우저 여는 것은 GET

##### post: 리소스를 생성하거나 변경
* server에게 여러 개를 전송을 해도 응답은 다르게 올 수 있음
* 게시글을 작성하거나 삭제할 때는 post로 요청하면 서버에서 변경되도록 햊ㅁ
* 메세지 바디 
* uri=url+query string 

##### 응답메세지 = 응답헤더 +응답 바디
* 응답헤더: HTTP버전, 상태 코드, 일시, 콘텐츠 형태, 인코딩 방식, 크기 등이 포함됨    
    상태 코드는 다섯개로 나눠짐   
       
* 응답바디: HTML이 포함됨.

##### mime(multipurpose internet mail Extensions)
- 다목적 인터넷 메일 확장
- 웹을 통해 여러 형태의 파일을 전달할 때 사용

##### ascii 인코딩
- 영문자를 컴퓨터가 해석할 수 있는 숫자로 변환시켜줌
    - 1byte(8bit)중에서 7bit를 이용하여 총 127개의 문자를 표현할 수 있음
        * bit=1,0으로 되어있는 것

## 웹크롤러?
* 웹문서 이미지 등을 주기적으로 수집하여 자동으로 데이터베이스화 시스템을 만드는 것 

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("https://www.sedaily.com/stock/005930")
bsObject = BeautifulSoup(html,"html.parser")
```

```python
print(bsObject)
```

```python
for meta in bsObject.head.find_all('meta'):
    print(meta.get('content'))
```

```python
# meta tag를 가져와서 get함수를 이용하여 속성 이름이 content이고 그에 해당하는 내용을 가져옴
# find함수는 태그명과 속성명 모두 가져오도록 할 수 있음
# key가 content인 value출력
print(bsObject.head.find("meta",{'name':'Keywords'}).get('content'))
```

```python
# 링크 가져오기
# a tag로 둘러싸인 text와 href속성 출력
# strip: 앞뒤 공백 삭제 
for link in bsObject.find_all('a'):
    print(link.text.strip(),link.get('href'))
```
