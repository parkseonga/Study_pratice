```python
from lxml import etree
import xml.etree.ElementTree as ElementTree
import pandas as pd
```

```python
from urllib.request import urlopen
from urllib.parse import urlencode,unquote,quote_plus
import urllib
```

```python
# 빈 벡터 생서
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
    

for i in range(1,137):
    
    # api주소 열기 (url + 인증키 + 해당페이지+한 화면에 출력되는 데이터 개수)
    url = 'http://api.kiscon.net/openapi-data/service/rest/kiscon/getGongsiRegList'
    queryParams = '?' + 'serviceKey='+'QgPOo6zCRMpwFdtdiSrZ5mkZXCFH4N24opuVNsfkiKafNis38QkEShEB7C%2Fu09m1ukguAbfTHJL%2BeV62phyhzg%3D%3D'\
    +'&pageNo='+str(i)\
    +'&numOfRows=1000'+'&sDate='+'20030101'+'&eDate='+'20191111'    
    
    request = urllib.request.Request(url+queryParams)
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()


    # 추출된 xml형식의 text를 xml객체로 파싱

    tree = etree.fromstring(response_body)

    # 태그로부터 원하는 텍스트 추출
    for media in tree.getiterator('item'):
        a.append(media.findtext('ncrGsAddr'))
        b.append(media.findtext('ncrGsDate'))
        c.append(media.findtext('ncrGsFlag'))
        d.append(media.findtext('ncrGsKname')) 
        e.append(media.findtext('ncrGsMaster'))
        f.append(media.findtext('ncrGsReason'))
        g.append(media.findtext('ncrItemName'))
        h.append(media.findtext('ncrItemregno'))
        
    
```
